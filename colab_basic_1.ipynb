```python
import matplotlib.pyplot as plt

# 학생 데이터 입력 (이름, 국어, 수학, 과학)
students = []

for i in range(5):
    print(f"학생 {i+1} 정보 입력:")
    name = input("이름: ")
    korean = int(input("국어 점수: "))
    math = int(input("수학 점수: "))
    science = int(input("과학 점수: "))
    students.append({'이름': name, '국어': korean, '수학': math, '과학': science})

print("\n=== 학생별 성적 및 합계/평균 ===")
names = []
totals = []
avgs = []

for s in students:
    total = s['국어'] + s['수학'] + s['과학']
    avg = total / 3
    print(f"{s['이름']} - 합계: {total}, 평균: {avg:.2f}")
    names.append(s['이름'])
    totals.append(total)
    avgs.append(avg)

# 전체 과목별 평균 구하기
total_korean = sum(s['국어'] for s in students)
total_math = sum(s['수학'] for s in students)
total_science = sum(s['과학'] for s in students)

print("\n=== 전체 과목별 평균 ===")
print(f"국어 평균: {total_korean / 5:.2f}")
print(f"수학 평균: {total_math / 5:.2f}")
print(f"과학 평균: {total_science / 5:.2f}")

# 시각화

# 1. 학생별 합계 및 평균 그래프
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.bar(names, totals, color='skyblue', label='합계')
plt.bar(names, avgs, color='orange', label='평균', alpha=0.7)
plt.title('학생별 합계 및 평균')
plt.xlabel('학생 이름')
plt.ylabel('점수')
plt.legend()

# 2. 과목별 평균 그래프
subjects = ['국어', '수학', '과학']
subject_avgs = [total_korean/5, total_math/5, total_science/5]

plt.subplot(1,2,2)
plt.bar(subjects, subject_avgs, color=['red', 'green', 'blue'])
plt.title('과목별 평균 점수')
plt.xlabel('과목')
plt.ylabel('평균 점수')
plt.ylim(0, 100)

plt.tight_layout()
plt.show()
```
